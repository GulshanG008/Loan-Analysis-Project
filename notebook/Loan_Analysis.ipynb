```python
import pandas as pd

df = pd.read_csv("../data/raw/loan_dataset.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
sns.countplot(x='loan_approved', data=df)
plt.title("Loan Approval Distribution")
plt.show()
```

```python
sns.boxplot(x='loan_approved', y='income',data=df)
plt.title("Income vs Loan Approval")
plt.show()
```

```python
sns.boxplot(x='loan_approved', y='credit_score',data=df)
plt.title("Credit Score vs Loan Approval")
plt.show()
```

```python
sns.boxplot(x='loan_approved', y='dependents',data=df)
plt.title("Dependents vs Loan Approval")
plt.show()
```

```python
sns.countplot(x='home_owner', hue='loan_approved',data=df)
plt.title("Home Ownership vs Loan Approval")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True,
            cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
